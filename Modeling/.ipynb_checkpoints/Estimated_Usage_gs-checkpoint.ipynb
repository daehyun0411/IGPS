```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import warnings


plt.rcParams['axes.unicode_minus'] = False
plt.rcParams["figure.figsize"] = (16,8)
plt.rc('font', family='Malgun Gothic')
warnings.filterwarnings('ignore')

%matplotlib inline
%config InlineBackend.figure_format = 'retina'
```

# 인천시 데이터

```python
gs = pd.read_csv("C:/Users/keogu/Desktop/IGPS/Data/강서구_대여소정보.csv")
```

```python
gs.drop(['field_1'],axis=1,inplace=True)
gs.drop(columns=['fid'], inplace=True)
```

```python
gs
```

```python
gs = gs[gs['강서_그리드_우선순위_rank'] < 84]
```

```python
gs.set_index(['강서_그리드_우선순위_rank'], inplace=True)
```

```python
# 분포확인
plt.figure(figsize=(16,24)) 
n=1
for feature in gs.columns:
    ax = plt.subplot(4,4,n) 
    plt.title('%s의 분포' %feature, fontsize=15)
    sns.distplot(gs[feature], kde =False)
    plt.xticks(rotation=90)
    n+=1
plt.tight_layout() 
plt.show()
```

```python
# 왜도 확인
from scipy.stats import skew

features_index = gs.dtypes[gs.dtypes != 'object'].index
skew_features = gs[features_index].apply(lambda x : skew(x))
skew_features_top = skew_features[skew_features > 1]
print(skew_features_top.sort_values(ascending=False))
```

```python
# 복사
gs_log = gs.copy()
```

```python
# log변환
gs_log[skew_features_top.index] = np.log1p(gs[skew_features_top.index])
gs_log
```

# 서울시 데이터

```python
total = pd.read_csv("C:/Users/keogu/Desktop/IGPS/Data/서울시_데이터종합_수정.csv")
```

```python
total
```

```python
total.set_index(['대여소번호'], inplace=True)
```

```python
# 왜도 확인
from scipy.stats import skew

features_index = total.dtypes[total.dtypes != 'object'].index
skew_features = total[features_index].apply(lambda x : skew(x))
skew_features_top = skew_features[skew_features > 1]
print(skew_features_top.sort_values(ascending=False))
```

```python
# 복사
total_log = total.copy()
```

```python
# log변환
total_log[skew_features_top.index] = np.log1p(total[skew_features_top.index])
total_log
```

# 회귀분석

```python
y_train = total_log['일별대여평균이용량']
X_train = total_log.drop('일별대여평균이용량', axis=1, inplace=False)
```

```python
X_test = gs_log
```

```python
y_train_scaled = y_train.copy()
X_train_scaled = X_train.copy()
X_test_scaled = X_test.copy()
```

```python
from sklearn.preprocessing import MinMaxScaler

#원래 값을 구하기 위해?
scaler_train = MinMaxScaler()
X_train_scaled = scaler_train.fit_transform(X_train)

scaler_test = MinMaxScaler()
X_test_scaled = scaler_test.fit_transform(X_test)

scaler_y = MinMaxScaler()
y_train_scaled = scaler_y.fit_transform(y_train.values.reshape(-1,1))
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X_train_scaled, y_train_scaled)
pred_lr = lr.predict(X_test_scaled)
```

```python
pred_lr= scaler_y.inverse_transform(np.array(pred_lr).reshape(-1,1))
pred_lr = np.expm1(pred_lr)
```
