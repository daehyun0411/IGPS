## 토지이용혼합도

```python
from numpy import log as ln
```

```python
ich['sum'] =  ich.iloc[:,2] + ich.iloc[:,3] + ich.iloc[:,4] + ich.iloc[:,5] + ich.iloc[:,6] + ich.iloc[:,7] + ich.iloc[:,8] + ich.iloc[:,9]
```

```python
for i in range(2,10):
    ich.iloc[:,i] = -1 * (ich.iloc[:,i]/ich['sum']) * ln(ich.iloc[:,i]/ich['sum'])
ich.fillna(0, inplace=True)
```

```python
ich['LUM'] = (ich.iloc[:,2] + ich.iloc[:,3] + ich.iloc[:,4] + ich.iloc[:,5] + ich.iloc[:,6] + ich.iloc[:,7] + ich.iloc[:,8] + ich.iloc[:,9])/ln(8)
```
